```python
import cv2
import numpy as np

def generate_chessboard(rows, cols, square_size):
    chessboard = np.zeros((rows * square_size, cols * square_size), dtype=np.uint8)

    for i in range(rows):
        for j in range(cols):
            if (i + j) % 2 == 0:
                chessboard[i * square_size:(i + 1) * square_size, j * square_size:(j + 1) * square_size] = 255

    return chessboard

# Specify the size of the chessboard
rows = 8
cols = 8
square_size = 50  # Size of each square in pixels

# Generate the chessboard pattern
chessboard_pattern = generate_chessboard(rows, cols, square_size)

# Display the generated chessboard pattern
cv2_imshow( chessboard_pattern)
```

```python
import cv2
import numpy as np

def generate_euclidean_distance_pattern(rows, cols):
    center = (rows // 2, cols // 2)
    euclidean_pattern = np.zeros((rows, cols), dtype=np.uint8)

    for i in range(rows):
        for j in range(cols):
            distance = np.sqrt((i - center[0])**2 + (j - center[1])**2)
            euclidean_pattern[i, j] = int(distance)

    return euclidean_pattern

# Specify the size of the image
rows = 256
cols = 256

# Generate the Euclidean distance pattern
euclidean_pattern = generate_euclidean_distance_pattern(rows, cols)

# Display the generated pattern
cv2_imshow( euclidean_pattern)

```

```python
import cv2
import numpy as np

def generate_diagonal_pattern(rows, cols):
    diagonal_pattern = np.zeros((rows, cols), dtype=np.uint8)

    for i in range(rows):
        for j in range(cols):
            if i < j:
                diagonal_pattern[i, j] = 255

    return diagonal_pattern

# Specify the size of the image
rows = 256
cols = 256

# Generate the diagonal pattern
diagonal_pattern = generate_diagonal_pattern(rows, cols)

# Display the generated pattern
cv2.imshow("Diagonal Pattern", diagonal_pattern)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# prompt: give me code using im show in image processing for city block pattern using matrices

import numpy as np
import matplotlib.pyplot as plt

# Create a 256x256 image with a black background
image = np.zeros((256, 256), dtype=np.uint8)

# Generate the city block pattern
city_block_pattern = np.zeros((256, 256), dtype=np.uint8)

for i in range(256):
    for j in range(256):
        city_block_pattern[i, j] = abs(i - 128) + abs(j - 128)

# Copy the city block pattern onto the image
image[:] = city_block_pattern

# Display the image using imshow
plt.imshow(image, cmap='gray')
plt.show()
```

```python
import cv2
import numpy as np

def blend_images(image1_path, image2_path, alpha, output_path):
    # Read the two images
    image1 = cv2.imread(image1_path)
    image2 = cv2.imread(image2_path)

    if image1 is None or image2 is None:
        print("Error: Unable to read one or both images.")
        return

    # Make sure both images have the same size
    if image1.shape != image2.shape:
        print("Error: Images must have the same dimensions.")
        return

    # Blend the two images using addWeighted
    result = cv2.addWeighted(image1, alpha, image2, 1 - alpha, 0)

    # Display or save the result
    cv2.imshow("Blended Images", result)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    # Save the result image
    cv2.imwrite(output_path, result)
    print(f"Result saved to {output_path}")

# Paths to the two images and the output image
image1_path = "/content/im1.jpg"  # Replace with the actual path
image2_path = "/content/im2.jpg"  # Replace with the actual path

alpha = 0.5  # Adjust the blending ratio, where 0.0 means 100% image1 and 1.0 means 100% image2

# Call the function to blend the two images
blend_images(image1_path, image2_path, alpha)
```
